Making predictions
Predicted values of a fitted estimator are the estimated output values for values of the input features. Predicted values are denoted as yhat, pronounced "y-hat". The .predict() method makes predictions from a fitted estimator. Ex: pred = model.predict(X) assigns the predictions from model for the values of X to pred. Predicted values can be calculated for the original inputs or for new data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset and drop instances with missing values
rides = pd.read_csv("cab_rides.csv").dropna()

# X = dataframe of input features
X = rides[['distance']]

# y = dataframe of the output feature
y = rides[['price']]
```

```python
# Initialize the model
linearModel = LinearRegression()

# Fit the model
linearModel.fit(X, y)
```

```python
# Make predictions for X
linearModel.predict(X)
```

```python
# Make predictions for a new instance
# distance = 2.0 miles
Xnew = pd.DataFrame({'distance': [5.0]})
print(Xnew)
```

```python
linearModel.predict(Xnew)
```
