Fitting an estimator
Once an estimator has been initialized, the .fit() method estimates the model weights or applies the algorithm to the input and output features. .fit() works with any scikit-learn estimator and takes two values: X and y. model.fit(X, y) updates the model object with the fitted estimator.

For models with coefficients, such as linear regression, the fitted estimator contains attributes coef_ and intercept_. model.coef_ contains the fitted weights, and model.intercept_ contains the fitted intercept (if applicable).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset and drop instances with missing values
rides = pd.read_csv("cab_rides.csv").dropna()
```

```python
# X = dataframe of input features
X = rides[['distance']]

# y = dataframe of the output feature
y = rides[['price']]
```

```python
# Initialize the model
linearModel = LinearRegression()
```

```python
# Fit the model
#The method .fit() fits the initialized model to input feature(s) X and output feature y.
linearModel.fit(X, y)
```

```python
# Intercept
#For models with weights, like logistic regression, logModel.intercept_, returns the y-intercept, w0.
linearModel.intercept_
```

```python
# Weight
#logModel.coef_ returns the slope of the logistic regression model, w1.
linearModel.coef_
```
