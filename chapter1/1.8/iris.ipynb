```python
# Import the necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression

# Read in the file fg_attempt.csv
iris_dataset = pd.read_csv('iris_data.csv')

print (f"{iris_dataset.shape}")
```

```python
x1Name = 'petal_length'
x2Name = 'petal_width'
x3Name = 'sepal_length'
x4Name = 'sepal_width'
iris_dataset.head()
```

```python
# Create a dataframe X containing 2 features
X = iris_dataset[[x1Name,x2Name]]
X.head()
```

```python
# Create a dataframe y containing 'class'
y = iris_dataset[['class']]
y.head()
```

```python
p = sns.scatterplot(data=iris_dataset, x=x1Name, y=x2Name, hue='class', style='class')
```

```python
p2 = sns.scatterplot(data=iris_dataset, x=x3Name, y=x2Name, hue='class', style='class')
```

```python
# Flatten y into an array
y = np.ravel(y)
```

```python
# Initialize a LogisticRegression() model
logisticModel = LogisticRegression()
```

```python
# Fit the model
logisticModel.fit(X, y)
```

```python
# Input feature values for a sample instance
x1 = float(input())
x2 = float(input())
```

```python
XNew = pd.DataFrame({'petal_length':[x1], 'petal_width':[x2]})
```

```python
# Predict the outcome from the new data
pred = logisticModel.predict(XNew)
print(pred)
```

```python
# Determine the accuracy of the model logisticModel
score = logisticModel.score(X, y)
print(f"{score:.2f}")
```
