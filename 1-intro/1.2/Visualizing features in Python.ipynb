Plotting features in Python
Exploratory data analysis is the process of exploring data via plots and descriptive statistics to identify potential relationships and patterns. Two popular Python libraries for plotting data are matplotlib and seaborn.

-matplotlib is a library for creating static, dynamic, and interactive plots.
-seaborn is a high-level library for creating plots that work especially well with dataframes.
Since seaborn is based on matplotlib, seaborn plots can be customized using functions and color palettes from matplotlib. The seaborn documentation and matplotlib documentation are handy references for creating and modifying plots.
The basic plot syntax in seaborn is sns.plot_type(data=df, x='x', y='y'). Each plot type has a unique seaborn function. Ex: sns.scatterplot() creates a scatter plot. Features are assigned to the x-axis and y-axis using the x and y parameters, respectively. Additional parameters hue and style change the plot color or marker shape based on values of a feature.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Using a random sample of rides speeds up visualization
rides = pd.read_csv('cab_rides.csv')
rides = rides.sample(1000, random_state=123)
```

```python
rides.head()
```

```python
# sns.countplot() creates a bar plot
plt.figure(figsize=(16,12))
p = sns.countplot(data=rides, x='destination', hue='destination')

# matplotlib functions are used to customize labels
p.set_xlabel('Destination', fontsize=14)
p.set_ylabel('Count', fontsize=14)
```

```python
# sns.histplot() creates a histogram
p = sns.histplot(data=rides, x='price', bins=20)
p.set_xlabel('price', fontsize=14)
p.set_ylabel('Count', fontsize=14)
```

```python
# sns.scatterplot() creates a scatter plot
p = sns.scatterplot(data=rides, x='distance', y='price', hue='cab_type', style='cab_type')
p.set_xlabel('Distance (miles)', fontsize=14)
p.set_ylabel('Price', fontsize=14)
p.legend(title='Cab type')
```
