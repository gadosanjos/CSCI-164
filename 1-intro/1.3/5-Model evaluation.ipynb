Evaluating estimators
Model evaluation metrics are numerical measures of how well an estimator's predicted values compare against the output feature's observed values. The two most commonly used metrics are accuracy and R-squared.

Accuracy is the proportion of instances correctly classified in a classification model. Accuracy ranges from 0.0 to 1.0, with higher accuracy indicating a better estimator.

R-squared measures the approximate proportion of variation in the output feature explained using a regression model. The maximum value of R-squared is 1.0, which means the estimator perfectly predicted the output feature. Negative values of R-squared mean that the estimator predicted worse than predicting the mean output for all instances.

scikit-learn estimators return evaluation metrics using the .score() method. Classification models will return accuracy, and regression models will return R-squared.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset and drop instances with missing values
rides = pd.read_csv("cab_rides.csv").dropna()

# X = dataframe of input features
X = rides[['distance']]

# y = dataframe of the output feature
y = rides[['price']]
```

```python
# Initialize the model
linearModel = LinearRegression()

# Fit the model
linearModel.fit(X, y)

# Make predictions
pred = linearModel.predict(X)
```

```python
# Score = R-squared
#The model explains 0.102 or 10.2% of the variation in price
#R-squared measures the proportion of variation in the output feature, price, explained by the model. 
# A trip's distance is not enough to explain all changes in price.
linearModel.score(X, y)
```

```python
# Weak relationship between actual price and predicted price
p = sns.scatterplot(x=np.ravel(y), y=np.ravel(pred))
p.set_xlabel('Actual price', fontsize=14)
p.set_ylabel('Predicted price', fontsize=14)
plt.xlim(0, 100)
plt.ylim(0, 100)
```

```python
ActualPrice = rides.loc[500:510,['distance','price']]
Xnew = pd.DataFrame({'distance': [3.14]})
predictedPrice = linearModel.predict(Xnew)
print(ActualPrice)
print(predictedPrice)
```
