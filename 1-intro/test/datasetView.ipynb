```python
rides = pd.read_csv('cab_rides.csv')
rides = rides.sample(1000, random_state=123)
# List of all columns
cols = rides.columns.tolist()
print(cols)
```

1. Count Plots
Description:
Count plots, also known as bar plots, are used to visualize the frequency of categorical data. They display the number of observations in each category and are particularly useful for highlighting the distribution of categorical features.

Best Used For:
- Categorical Data: Variables that represent distinct groups or categories.
- Small to Medium Number of Categories: Too many categories can make the plot cluttered and hard to read.
- Visualizing Counts: Quickly shows which categories are most or least frequent.
When to Use:
- Comparing the Frequency of Categories: If you want to see how many observations fall into each category.
- Exploring Distribution of Non-Numeric Data: Useful in initial data exploration to understand categorical variables.
- Identifying Trends or Patterns in Categorical Data: Can highlight the most common or rare categories.
Why Itâ€™s Effective:
- Clear Representation: Easy to interpret and compare the counts of different categories.
- Highlighting Differences: Helps in spotting the categories with the most or least counts at a glance.
- Color Coding: Can use different hues to compare additional categorical dimensions within the same plot.
Example:
- Visualizing Cab Types: Showing the number of rides for each cab type in a dataset.

```python
# Count Plots
# Use count plots for categorical data to show the distribution of categories.
# Count Plots for Categorical Columns
for x in cols:
    if pd.api.types.is_object_dtype(rides[x]) or pd.api.types.is_categorical_dtype(rides[x]):
        plt.figure(figsize=(16,12))  # Specified size
        sns.countplot(data=rides, x=x, hue=x)  # Create the count plot
        plt.title(f'Count Plot for {x}')  # Add a title
        plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
        plt.tight_layout()  # Adjust layout
        plt.show()  # Display the plot
```

Histograms
Description:
Histograms are used to represent the distribution of a numeric variable. They display the frequency of data within certain range intervals (bins). Histograms give a visual interpretation of the numerical data by showing the number of data points that fall within specified ranges.

Best Used For:
- Numeric Data: Variables that are quantitative and continuous.
- Understanding Distribution: Showing how data points are distributed across different ranges.
- Large Datasets: Effective for large amounts of data to show the underlying frequency distribution.
When to Use:
- Analyzing Distribution Shape: To see if data follows a normal distribution, is skewed, or has multiple peaks.
- Identifying Central Tendency and Spread: To understand the typical values and variability in the data.
- Spotting Outliers: Outliers may appear as bars that are isolated from the rest.
Why Itâ€™s Effective:
- Frequency Distribution: Provides a clear picture of how data is distributed across different intervals.
- Identifying Patterns: Helps in recognizing patterns like skewness or bimodality.
- Statistical Analysis: Aids in determining the statistical properties of the data, such as mean, median, and standard deviation.
Example:
- Price Distribution: Showing how frequently different ride prices occur in a dataset.

```python
# Histograms
# Use histograms for numeric columns to show the distribution of numerical values.
# Histograms for Numeric Columns
for x in cols:
    if pd.api.types.is_numeric_dtype(rides[x]):  # Check if the column is numeric
        plt.figure(figsize=(10, 6))  # Create a new figure
        sns.histplot(data=rides, x=x, kde=True)  # Create the histogram with a kernel density estimate
        plt.title(f'Histogram for {x}')  # Add a title
        plt.xlabel(x)  # Label for x-axis
        plt.ylabel('Frequency')  # Label for y-axis
        plt.tight_layout()  # Adjust layout
        plt.show()  # Display the plot
```

Scatter Plots
Description:
Scatter plots display values for two variables for a set of data. They plot data points on a Cartesian plane, where each point represents the values of two numeric variables. Scatter plots are great for visualizing the relationship between two numeric variables.

Best Used For:
- Numeric Data: Variables that are quantitative and continuous.
- Examining Relationships: Useful for identifying correlations or patterns between two variables.
- Detecting Outliers: Points that fall far from the rest can be easily spotted.
When to Use:
- Exploring Relationships: To see if changes in one variable correspond with changes in another.
- Identifying Trends: To observe linear or nonlinear trends.
- Assessing Variability: To see how spread out the data points are.
Why Itâ€™s Effective:
- Visual Correlation: Helps in assessing how one variable may affect another (e.g., positive, negative, or no correlation)
- Trend Identification: Easily identify patterns or clusters in data.
- Outlier Detection: Outliers stand out clearly, which can be crucial for data quality and analysis.
Example:
- Distance vs. Price: Showing how ride distance correlates with the ride price in a dataset.

```python
# scatter plots
# Use scatter plots for numeric columns to show relationships between numeric pairs.
# Identify numeric columns
numeric_cols = rides.select_dtypes(include='number').columns.tolist()

for x in numeric_cols:
    for y in numeric_cols:
        if x != y:  # Avoid plotting the same variable against itself
            plt.figure(figsize=(10, 6))  # Create a new figure
            sns.scatterplot(data=rides, x=x, y=y)  # Create the scatter plot
            plt.title(f'Scatter Plot for {x} vs {y}')  # Add a title
            plt.xlabel(x)  # Label for x-axis
            plt.ylabel(y)  # Label for y-axis
            plt.tight_layout()  # Adjust layout
            plt.show()  # Display the plot
```
